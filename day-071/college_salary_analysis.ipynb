```python
clean_df = df.dropna()
clean_df.tail()
```

```python
clean_df['Starting Median Salary'].head()
```

Highest salary

```python
clean_df['Starting Median Salary'].max()
```

College with highest starting salary

```python
id_college_highest_starting_salary = clean_df['Starting Median Salary'].idxmax()
clean_df.loc[id_college_highest_starting_salary, ['Undergraduate Major', 'Starting Median Salary']]
```

College with lowest starting salary

```python
id_college_lowest_starting_salary = clean_df['Starting Median Salary'].idxmin()
clean_df.loc[id_college_lowest_starting_salary, ['Undergraduate Major', 'Starting Median Salary']]
```

College with highest salary in mid-career

```python
id_college_highest_mid_career_salary = clean_df['Mid-Career Median Salary'].idxmax()
clean_df.loc[id_college_highest_mid_career_salary, ['Undergraduate Major', 'Mid-Career Median Salary']]
```

College with highest salary in mid-career

```python
id_college_lowest_salary_mid_career = clean_df['Mid-Career Median Salary'].idxmin()
clean_df.loc[id_college_lowest_salary_mid_career]
```

Lowest Risk Majors

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

Sorting by the Lowest Spread

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

Majors with the Highest Potential

```python
columns = ['Undergraduate Major', 'Mid-Career 90th Percentile Salary']
clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)[columns].head()
```

Majors with the Greatest Spread in Salaries

```python
columns = ['Undergraduate Major', 'Spread']
clean_df.sort_values('Spread', ascending=False)[columns].head()
```

Average salary per group

```python
clean_df.groupby('Group').mean(numeric_only=True).round(2)
```
